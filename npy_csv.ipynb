```python
import pandas as pd
prefix_face = "face"
face_coor = 80
prefix_pose = "pose"
pose_coor = 33
prefix_left_hand = "left_hand"
prefix_right_hand = "right_hand"
hand_coor = 21

framenumber  = 0
coor_per_frame = 155
maximumframe = int(targetlength/465)
maximumframe
```

```python
df = pd.DataFrame()
counter = 0
for i in range(maximumframe):
    #face
    counter = 0
    column_name_x=f'{prefix_face}{i}_x'
    column_name_y=f'{prefix_face}{i}_y'
    column_name_z=f'{prefix_face}{i}_z'
    df[column_name_x]=None
    df[column_name_y]=None
    df[column_name_z]=None
    for j in range(i*coor_per_frame,i*coor_per_frame+face_coor):
        df.loc[counter, column_name_x] = reshaped_array[j,0]
        df.loc[counter, column_name_y] = reshaped_array[j,1]
        df.loc[counter, column_name_z] = reshaped_array[j,2]
        counter=counter+1

    #pose
    column_name_x=f'{prefix_pose}{i}_x'
    column_name_y=f'{prefix_pose}{i}_y'
    column_name_z=f'{prefix_pose}{i}_z'
    df[column_name_x]=None
    df[column_name_y]=None
    df[column_name_z]=None
    counter = 0
    for k in range(i*coor_per_frame+face_coor,i*coor_per_frame+face_coor+pose_coor):
        df.loc[counter, column_name_x] = reshaped_array[k,0]
        df.loc[counter, column_name_y] = reshaped_array[k,1]
        df.loc[counter, column_name_z] = reshaped_array[k,2]
        counter=counter+1
        
    #left hand
    column_name_x=f'{prefix_left_hand}{i}_x'
    column_name_y=f'{prefix_left_hand}{i}_y'
    column_name_z=f'{prefix_left_hand}{i}_z'
    df[column_name_x]=None
    df[column_name_y]=None
    df[column_name_z]=None
    counter = 0
    for m in range(i*coor_per_frame+face_coor+pose_coor,i*coor_per_frame+face_coor+pose_coor+hand_coor):
        df.loc[counter, column_name_x] = reshaped_array[m,0]
        df.loc[counter, column_name_y] = reshaped_array[m,1]
        df.loc[counter, column_name_z] = reshaped_array[m,2]
        counter=counter+1

    #right hand
    column_name_x=f'{prefix_right_hand}{i}_x'
    column_name_y=f'{prefix_right_hand}{i}_y'
    column_name_z=f'{prefix_right_hand}{i}_z'
    df[column_name_x]=None
    df[column_name_y]=None
    df[column_name_z]=None
    counter = 0
    for n in range(i*coor_per_frame+face_coor+pose_coor+hand_coor,i*coor_per_frame+face_coor+pose_coor+hand_coor+hand_coor):
        df.loc[counter, column_name_x] = reshaped_array[n,0]
        df.loc[counter, column_name_y] = reshaped_array[n,1]
        df.loc[counter, column_name_z] = reshaped_array[n,2]
        counter=counter+1
        
        
#print(df["face0_x"])
df.to_csv('output.csv', index=False)
```
